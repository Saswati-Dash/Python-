```python
# Offers generator for walmart.
# 20% --> customer --> 1500 or 3 or more transaction with walmart
# task : Final amount customer needs to pay

Total_amount = float(input('Add the bill amount '))
Repeat_pur = int(input('Enter the Frequency of purchase'))
Membership = input('Enter the Membership status ')

if Total_amount >1500 or Repeat_pur>=3 or Membership == 'Yes':
  print('Amount to be paid', Total_amount*0.8)

else :
  print('Amount to be paid',Total_amount)

```

```python
# Offers generator for walmart.

# 20% --> customer --> 1500 or 3 or more transaction
# 40% --> Gold Member
# 17% ---> Customer--> 1000
# 36% --> Platnium

trans_amount = float(input('Enter the Amount '))
repeat_purchase = int(input('Number of time doing transation in a month '))
memb = input('Enter the Membership type gold or plat or no ')


if memb=='gold':
  print('Amount to be paid ',trans_amount*0.6)
elif memb=='plat':
  print('Amount to be paid ',trans_amount*0.64)
elif trans_amount>1500 or repeat_purchase>=3:
  print('Amount to be paid ',trans_amount*0.8)
elif trans_amount>1000 :
  print('Amount to be paid ',trans_amount*0.83)
else :
  print('Amount to be paid ',trans_amount)

```
