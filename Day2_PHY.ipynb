```python
annual_ctc=float((input("Enter your annual CTC: ")))
tax = 0.1*annual_ctc
pf = 0.05*annual_ctc
annual_in_hand= annual_ctc-tax-pf
monthly_in_hand = annual_in_hand/12
print("monthly salary",monthly_in_hand)
print("Total Tax",tax)
print("PF deduction" , pf)
```
