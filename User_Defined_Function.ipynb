```python
# Write a loop that counts number of vowels in a word:

word = input("Write a word : ")
vowels = "aeiouAEIOU"
count = 0
for i in word:
  if i in vowels:
    count += 1
print(f"{word} has {count} vowels")
```

```python
# Write a for Loop that will positionaly multiply number in two lists
L1 = [2,3,4]
L2 = [4,5,6]
L3 = []
for i in range (len(L1)):
  L3.append(L1[i]*L2[i])
print(L3)
```

```python
# Enumerate - Produces two output - Index, Element
l = ["Jasmine", "Harry", "Radhika" , "Rahul"]
for i,j in enumerate(l):
  print(i,j)
```

```python
## User Defined Functions
# 1. These are those functions that one can customise in order to meet a very niche objective
# 2. They are named/known functions
# 3. They are reusable and readable
# 4. We can write multiple lines of code of user defined  function
# 5. They support return keyword
# 6. they are defined using def keyword


```

```python
# Write a user defined function that adds two numbers
def add(x,y):
  result = x+y
  return result
```

```python
# Create a user defined function that will multiply 3 number
def mul(x,y,z):
  result = x*y*z
  return result
```

```python
mul(2,3,4)
```

```python
a = mul(2,3,5)
```

```python
# Exponent
def exponent(x,s):
  return x**s
```

```python
def square(c):
  return c*c
```

```python
# Write a function that prints odd for an odd number and even for an even number
def odd_even(x):
  if x%2==0:
    return f"{x} is an even number"
  else:
    return f"{x} is an odd number"
```

```python
def factorial(x):
  f = 1
  for i in range(1,x+1):
    f*=i
  return f
```

```python
factorial(5)
```

```python
# Create a user defined function that counts number of vowels in a word
def count_vowel(word):
  vowels = "aeiouAEIOU"
  count = 0
  for char in word:
    if char in vowels:
      count += 1
  return count
```

```python
count_vowel("apple")
```

```python
# Lamda functions
# They are similar to user defined functions in context to how the code is curated to meet a specific task but they are one liner function
# They are called unknown/anonymous function
# They do not support return keyword
# they are smaller and usually used for one time operations
# they are used using the lambda keyword
```

```python
# create a Lamda function that add two number
add = lambda x,y: x+y
```

```python
print(add(3,4))
```

```python
square = lambda x: x**2
```

```python
square(7)
```

```python
# Prints negative for a negative number , 0 for a 0 and positive for a positive number
assign = lambda x: "negative" if x<0 else "zero" if x==0 else "positive"
```

```python
assign(0)
```

```python
assign(4)
```

```python
assign(-5)
```
