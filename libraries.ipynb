```python
# %% is called a cell magic command .
# % - Works on single line . &  %% - Works on entire cell
# writefile - Writes the content of the cell into a file.

%%writefile libraries.py

def odd_even(x):
    if x % 2 == 0:
        return f"{x} Even"
    else:
        return f"{x} Odd"

def square(x):
    return x**2
```

```python
from libraries import odd_even, square

print(odd_even(10))
print(square(5))
```
