```python
c=[345,789,1090,2500,50,187]


# all product greater then
# >1500 - 32% discount
# >800 - 20%
# >300 - 15%
# >100 - 10%

for i in c:
  if i>1500:
    print(i*0.68)
  elif i>800:
    print(i*0.8)
  elif i>300:
    print(i*0.85)
  elif i>100:
    print(i*0.9)
  else:
    print(i)
```

```python
c=[345,789,1090,2500,50,187]

# blank list d_p is created and al the result are stored in d_p using append
d_p=[]

for i in c:
  if i>1500:
    d_p.append(i*0.68)
  elif i>800:
    d_p.append(i*0.8)
  elif i>300:
    d_p.append(i*0.85)
  elif i>100:
    d_p.append(i*0.9)
  else:
    d_p.append(i)
```

```python
# While Loop :- Conditional loop

# Looping - executes all the task till the condition remain true
# Termination --> conditions turns to be false
# instruction --> additional instruction is to be given to terminate while loop

# Print python in screen for 5 times
q = 1 # -- this is intiallization
while q<5:
  print('Python',q)
  q=q+1 # -- this is additional instruction

```

```python
 # Control Statement

 # Break : it will the terminate the loop as per the requirement
 # Continue : when condition meet it skip the task
 # Pass : used for creating up the empty logic section
```

```python
fly = ['Rose',58,96.34,True,78.25,35]
for a in fly :
  if a==96.34:
    break
  print(a)
```
