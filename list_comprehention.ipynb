```python
## List Comprehension
# 1. It is a short and concise way of creating lists in python
# 2. It can be used as a substitute for loops
# 3. They are generally faster and more efficent
```

```python
# Print all the number between 1 to 20
empty_list = []
for i in range(1,21):
  empty_list.append(i)
# here first we have to create an empty list then append output in it
```

```python
empty_list
```

```python
# but in list comprehension it is just a one line code
num_list = [i for i in range(1,21)]
```

```python
# write a for loop that will only print the even number
even_num = [i for i in range(1,21) if i%2==0]
```

```python
even_num
```

```python
Square = [i**2 for i in range (1,21)]
```

```python
Square
```

```python
# count the number of vowels in a word using list comprehension
word = "Aeroplane"
vowels = [sum(1 for i in word if i in "aeiouAEIOU")]
# here we want number
```
