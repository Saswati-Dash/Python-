```python
# Grading system :

# >90 --> A
#> 82--> B
# >75--> c
#>65 --> D
# >55 --> E
#<=55--> F

marks = float(input('Enter the marks out of 100'))

if marks>90:
  print('Student grade: A')
elif marks>82:
  print('Student grade: B')
elif marks>75:
  print('Student grade: C')
elif marks >65:
  print('Student grade: D')
elif marks>55:
  print('Student grade: E')
else:
  print('Student grade: F')
```

```python
# Classification of year as Leap year or non leap year .

#leap --> 366 days --> after every 4 year .

# Non leap --> 365 days .

# year%4==0 --> leap year .

#1900%4---> Non Leap year .

# Year ---> 2000,1900,2100,400,1700 divisible by 100 .
# centurian year --> divisible by 400 --> leap year --> not --> non leap year.

# Non centurian --> 1785,1789,2024,2026,2022,2020,1965.
# Divisible by 4 --> leap if not non leap year .
```

```python
year= int(input('Enter the Year'))

if year%400==0:
  print(year,'is a leap year')
elif year%100==0:
  print(year,'is non leap year')
elif year%4==0:
  print(year,'is a leap year')
else:
  print(year,'is non leap year')
```
